# Basic Classification of Fashion MNIST data

28 x 28 grayscale images of 10 types of clothes. More info on data: https://github.com/zalandoresearch/fashion-mnist/blob/master/README.md


```python
#Tensorflow and Keras
import tensorflow as tf
from tensorflow import keras

#Helper Libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
```

```python
#Loading the data
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
#Lets see an image
plt.figure()
plt.imshow(train_images[1])
plt.colorbar()
#plt.gca().grid(True)
```

```python
#We need to normalize these images
train_images = train_images / 255.0
test_images = test_images / 255.0

#Visualize the first 25 images
plt.figure(figsize=(10,10))

for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
```

## Build the model

```python
#The sequential model is a linear stack of layers
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(128, activation = tf.nn.relu),
                          keras.layers.Dense(10, activation=tf.nn.softmax)
                         ])

#Compile
model.compile(optimizer=tf.train.RMSPropOptimizer(learning_rate=0.1),
             loss='sparse_categorical_crossentropy',    #Cross-entropy loss increases as the predicted probability diverges from the actual label.
             metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5)
```

```python
predictions = model.predict(test_images)
```

## Plot the first 25 images
Color correct predictions in green, incorrect predictions in red

```python
plt.figure(figsize=(10, 10))

for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid("off")
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    predicted_label = np.argmax(predictions[i])
    true_label = test_labels[i]
    
    if predicted_label == true_label:
        color = 'green'
    else:
        color = 'red'
    plt.xlabel("{} ({})".format(class_names[predicted_label], class_names[true_label], color=color))
    
    
```
